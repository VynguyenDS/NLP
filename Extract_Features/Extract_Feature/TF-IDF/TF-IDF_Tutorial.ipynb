```python
import pandas as pd 
from sklearn.feature_extraction.text import TfidfVectorizer
documentA = 'the man went out for a walk'
documentB = 'the children sat around the fire'

```

```python
bagOfWordsA = documentA.split(' ')
bagOfWordsB = documentB.split(' ')
uniqueWords = set(bagOfWordsA).union(bagOfWordsB)
```

```python
numberOfWordA = dict.fromkeys(uniqueWords, 0)
for word in bagOfWordsA:
    numberOfWordA[word] += 1
    
```

```python
numOfWordsB = dict.fromkeys(uniqueWords, 0)
for word in bagOfWordsB:
    numOfWordsB[word] += 1
```

```python
countTokenInCorpus = pd.DataFrame(data = [numberOfWordA.values(),numOfWordsB.values()], columns =  numberOfWordA.keys(),index = ['A','B'])
countTokenInCorpus
```

```python
from tqdm import tqdm
def computeTF(wordDict, bagOfWords):
    tfDict = {}
    bagOfWordsCount = len(bagOfWords)
#     print(bagOfWordsCount)
    for word, count in tqdm(wordDict.items()):
        tfDict[word] = count / float(bagOfWordsCount)
    return tfDict
tfA = computeTF(numberOfWordA, bagOfWordsA)
tfA
```

```python
import math
def computeIDF(documents):
    n = len(documents)
    idfDict = dict.fromkeys(documents[0].keys(), 0)
   
    for document in tqdm(documents):
        for word, val in tqdm(document.items()):
            if val > 0:
                idfDict[word] += 1
    for word, val in idfDict.items():
        idfDict[word] = math.log(n/ float(val))
    return idfDict
    
```

```python
idfs = computeIDF([numberOfWordA, numOfWordsB])
idfs
```

```python
def computeTFIDF(tfBagOfWords, idfs):
    tfidf = {}
    for word, val in tfBagOfWords.items():
        tfidf[word] = val*idfs[word]
    return tfidf
```

```python
tfA = computeTF(numOfWordsA, bagOfWordsA)
tfB = computeTF(numOfWordsB, bagOfWordsB)

tfidfA = computeTFIDF(tfA, idfs)
tfidfB = computeTFIDF(tfB, idfs)
tfidfA = computeTFIDF(tfA, idfs)
tfidfB = computeTFIDF(tfB, idfs)
df = pd.DataFrame([tfidfA, tfidfB])
df
```
